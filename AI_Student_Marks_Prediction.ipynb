```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
study_hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
marks = np.array([35, 40, 50, 60, 70])

#train model
model = LinearRegression()
model.fit(study_hours, marks)
print("Model trained successfully!")
```

**Problem Statement**

The objective of this project is to build a simple AI model that predicts student marks based on the number of hours studied.

 **Dataset**

The dataset consists of two variables:

  1.Study Hours (input)

  2.Marks obtained (output)

 Sample data is used to demonstrate the working of the AI model.

```python
hours = [[6]]
predicted_marks = model.predict(hours)
print("Predicted marks for 6 hours study:", predicted_marks[0])
```

**Algorithm Used**

Linear Regression is used in this project. It is a supervised machine learning algorithm that learns the relationship between input and output values and predicts results for new data.

**Model Training**

The model is trained using study hours as input and marks as output. During training, the model learns a linear pattern between these values.

```python
plt.scatter(study_hours, marks, color='blue')
plt.plot(study_hours, model.predict(study_hours), color='red')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Student Marks Prediction using AI")
plt.show()
```

**Prediction**

After training, the model predicts the marks for a given number of study hours. For example, when the input is 6 study hours, the model predicts the expected marks.

**Result**

The graph shows actual data points and a straight line representing the prediction made by the AI model. This indicates that the model has learned the data pattern successfully.
